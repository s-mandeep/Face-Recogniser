# Reading an Image

```python
img=cv2.imread('dog.png')
## Default color mode is BGR when opencv reads

newImg=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
plt.show()
```

```python
plt.imshow(newImg)
plt.show()
```
